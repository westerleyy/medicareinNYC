```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('MEDICARE_PROVIDER_CHARGE_INPATIENT_DRGALL_FY2017.csv')
```

```python
# https://www.health.ny.gov/statistics/cancer/registry/appendix/neighborhoods.htm

zip_codes = ['104', '112', '100', '103']
# 104 Bronx
# 112 Brooklyn 
# 100 Manhattan
# 103 Staten Island

zip_codes_qn = ['110','111','113','114','116']
# these zipcodes for Queens
```

```python
df['Provider Zip Code'] = df['Provider Zip Code'].astype(str)
```

```python
df_ny = df[df['Provider State'] == 'NY']
```

```python
df_bx = df_ny[df_ny['Provider Zip Code'].str.startswith(zip_codes[0])]
df_bk = df_ny[df_ny['Provider Zip Code'].str.startswith(zip_codes[1])]
df_mn = df_ny[df_ny['Provider Zip Code'].str.startswith(zip_codes[2])]
df_si = df_ny[df_ny['Provider Zip Code'].str.startswith(zip_codes[3])]
df_qn = pd.DataFrame([])
df_nyc = pd.DataFrame([])
```

```python
for i in range(len(zip_codes_qn)):
    df_qn = df_qn.append(df_ny[df_ny['Provider Zip Code'].str.startswith(zip_codes_qn[i])])
```

```python
df_nyc = df_nyc.append([df_bx, df_bk, df_mn, df_si, df_qn])
```

```python
# Number of hospitals
bx_hosp = len(df_bx['Provider Name'].unique())
bk_hosp = len(df_bk['Provider Name'].unique())
mn_hosp = len(df_mn['Provider Name'].unique())
qn_hosp = len(df_qn['Provider Name'].unique())
si_hosp = len(df_si['Provider Name'].unique())

print('There are {} hospitals in The Bronx' .format(bx_hosp))
print('There are {} hospitals in Brooklyn' .format(bk_hosp))
print('There are {} hospitals in Manhattan' .format(mn_hosp))
print('There are {} hospitals in Queens' .format(qn_hosp))
print('There are {} hospitals in Staten Island' .format(si_hosp))
```

```python
# Number of hospitals per capita, population numbers from census.gov estimates for 2019
bx_pop = 1418207
bk_pop = 2559903
mn_pop = 1628706
qn_pop = 2253858
si_pop = 476143

print('There are {} hospitals/capita in The Bronx' .format(bx_hosp/bx_pop))
print('There are {} hospitals/capita in Brooklyn' .format(bk_hosp/bk_pop))
print('There are {} hospitals/capita in Manhattan' .format(mn_hosp/mn_pop))
print('There are {} hospitals/capita in Queens' .format(qn_hosp/qn_pop))
print('There are {} hospitals/capita in Staten Island' .format(si_hosp/si_pop))
```
