```python
import pandas as pd
import numpy as np
```

```python
tracts =pd.read_csv('ACSST5Y2018.S1901_data_with_overlays_2020-04-19T193840.csv', skiprows=[0])
tracts.head(3)
```

```python
tracts = tracts.join(pd.DataFrame(tracts['Geographic Area Name'].str.split(',').tolist(),
                                   columns = ['tract','county','city']))
tracts['tract'] = tracts['tract'].apply(lambda x : x[13:])
tracts.head(3)
```

```python
cols_to_keep = ['city', 'tract', 'county', 'Households (HH) Count', 'HH 0-10k', 'HH 10k-15k', 'HH 15k-25k', 
                'HH 25k-35k', 'HH 35k-50k', 'HH 50k-75k', 'HH 75k-100k', 'HH 100k-150k', 'HH 150k-200k', 'HH 200k-UP', 
                'Household Median Income', 'Household Mean Income']

prettier_cols = {
    'Estimate!!Households!!Total': 'Households (HH) Count',
    'Estimate!!Households!!Total!!Less than $10,000' : 'HH 0-10k',
    'Estimate!!Households!!Total!!$10,000 to $14,999': 'HH 10k-15k',
    'Estimate!!Households!!Total!!$15,000 to $24,999': 'HH 15k-25k',
    'Estimate!!Households!!Total!!$25,000 to $34,999': 'HH 25k-35k',
    'Estimate!!Households!!Total!!$35,000 to $49,999': 'HH 35k-50k',
    'Estimate!!Households!!Total!!$50,000 to $74,999': 'HH 50k-75k',
    'Estimate!!Households!!Total!!$75,000 to $99,999': 'HH 75k-100k',
    'Estimate!!Households!!Total!!$100,000 to $149,999': 'HH 100k-150k',
    'Estimate!!Households!!Total!!$150,000 to $199,999': 'HH 150k-200k',
    'Estimate!!Households!!Total!!$200,000 or more': 'HH 200k-UP',
    'Estimate!!Households!!Median income (dollars)': 'Household Median Income',
    'Estimate!!Households!!Mean income (dollars)': 'Household Mean Income',
}

list(tracts.columns)
```

```python
tracts.rename(columns=prettier_cols, inplace=True)
tracts.head()
```

```python
for_export = tracts[cols_to_keep]
for_export
```

```python
for_export.to_csv('household_income_acs2018.csv',index=False)
```
