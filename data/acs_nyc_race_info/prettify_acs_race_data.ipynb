```python
import pandas as pd
import numpy as np
```

```python
tracts =pd.read_csv('ACSDT5Y2018.B02001_data_with_overlays_2020-04-19T221609.csv', skiprows=[0])
tracts.head(3)
```

```python
tracts = tracts.join(pd.DataFrame(tracts['Geographic Area Name'].str.split(',').tolist(),
                                   columns = ['tract','county','city']))
tracts["county"].replace({" Bronx County": "Bronx", 
                          " Queens County": "Queens", 
                          " Kings County": "Brooklyn",
                          " Richmond County": "Staten Island",
                          " New York County": "Manhattan"}, inplace = True)
tracts['tract'] = tracts['tract'].apply(lambda x : x[13:])
tracts.head(3)
```

```python
tracts["BoroCTLbl"] = tracts["county"] + " " + tracts["tract"].astype(str)
```

```python
tracts.head()
```

```python
cols_to_keep = ['city', 'tract', 'county', "BoroCTLbl", 'Total', 'White', 'Black', 'Native', 'Asian', 
                'Pacific Islander', 'Other','Two or More']

prettier_cols = {
    'Estimate!!Total': 'Total',
    'Estimate!!Total!!White alone' : 'White',
     'Estimate!!Total!!Black or African American alone': 'Black',
     'Estimate!!Total!!American Indian and Alaska Native alone': 'Native',
     'Estimate!!Total!!Asian alone': 'Asian',
     'Estimate!!Total!!Native Hawaiian and Other Pacific Islander alone': 'Pacific Islander',
     'Estimate!!Total!!Some other race alone': 'Other',
     'Estimate!!Total!!Two or more races': 'Two or More'
}

list(tracts.columns)
```

```python
tracts.rename(columns=prettier_cols, inplace=True)
tracts.head()
```

```python
for_export = tracts[cols_to_keep]
for_export
```

```python
for_export.to_csv('tract_race_acs2018.csv',index=False)
```
